# DATA CLEANING

```python

from nltk.corpus import stopwords
import re
import nltk
from nltk import word_tokenize
```

```python

def clean_data(data):
    data=data.lower().replace('\n',' ').replace('\r','').strip()
    data=re.sub(' +',' ',data)
    data=re.sub(r'[^\w\s]','',data)
    stop_words=set(stopwords.words('english'))
    word_tokens = word_tokenize(data)
    data_pure=[]
    for i in word_tokens:
        if i not in stop_words:
            data_pure.append(i)
    data=' '.join(data_pure)
    return data
```

```python
dta['clean_text']=dta['text'].apply(clean_data)
```

```python
dta.tail()
```

# LABEL ENCODING

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
dta['category']=label_encoder.fit_transform(dta['category'])
```

# SPLITTING THE DATASET INTO TEST AND TRAIN DATASET

```python
from sklearn.model_selection import train_test_split
data_train,data_test,categ_train,categ_test=train_test_split(dta['clean_text'],dta['category'],test_size=0.3,random_state=8)
```

```python
data_test.shape
```

```python
data_train.shape
```

```python
categ_test.shape
```

```python
categ_train.shape
```

# CHANGING THE DATA INTO VECTOR

```python
from sklearn.feature_extraction.text import CountVectorizer
cv_obj_train=CountVectorizer()
cv_data_train=cv_obj_train.fit_transform(data_train)
cv_data_train.shape
```

```python
cv_data_train.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_obj_train=TfidfTransformer()
tfidf_data_train=tfidf_obj_train.fit_transform(cv_data_train)
tfidf_data_train.shape
```

```python

tfidf_data_train.toarray()
```

# IMPLEMENTING MULTINOMIAL NAIVE BAYES

```python
from sklearn.naive_bayes import MultinomialNB
clasifier=MultinomialNB()
clasifier.fit(tfidf_data_train,categ_train)
```

# TESTING THE OUTPUT ON NEW DATA SET

```python
cv_data_test=cv_obj_train.transform(data_test)
```

```python
tfidf_test_data=tfidf_obj_train.transform(cv_data_test)
```

```python
predicted_output=clasifier.predict(tfidf_test_data)
```

```python
predicted_output.shape
```

```python
data_test1=['tennis ball']
cv_data_test1=cv_obj_train.transform(data_test1)
tfidf_test_data1=tfidf_obj_train.transform(cv_data_test1)
predicted_output1=clasifier.predict(tfidf_test_data1)
```
